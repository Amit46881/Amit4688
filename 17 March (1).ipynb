```python
#Create a new object named Circle..
#take pie as a class Attribute , radius as a instance Attribute
#Define Methods to calculate area , create method to find circumference .. create a method to change the radius


class Circle():

    pie=3.14
    def __init__(self, radius):
        self.radius=radius

    def area(self):
        return self.pie*self.radius*self.radius

    def circumference(self):
        return 2*self.radius*self.pie

    def change_radius(self, newradius):
        self.radius=newradius
```

```python
y=Circle(10)
```

```python
y.radius
```

```python
y.area()
```

```python
y.circumference()
```

```python
y.change_radius(20)
```

```python
y.radius
```

```python
vars(y)
```

```python
#Inheritance
class Animal():

    def __init__(self):
        print("Animal Created")
    def who_am_i(self):
        print("I am animal")
    def eat(self):
        print("I love eating")
    
```

```python
#Create an object by the name Account, which takes owner_name and Balance as Attributes
#Create  methods of Withdrawl -- take input of withdrawl amount .. if the withdrael amt>balance give
#message as insuuficient funds otherwise update the balance and print the message .. wihdrawl Accepted
#Create another method of deposit.. update the balance .. and print the messahe deposit accepted

class Account():
    
    def __init__(self,Owner_name,balance):
        self.Owner_name=Owner_name
        self.balance=balance
    def deposit(self , amount):
        self.balance += amount
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawal accepted")
        else:
            print("Insuficient Funds")
```

```python
x=Account("Sam",10000)
```

```python
x.Owner_name
```

```python
x.deposit(1000)
```

```python
x.balance
```

```python
x.withdraw(5000)
```

```python
class Books():

    def __init__(self, title, author, pages):
        self.title=title
        self.author=author
        self.pages=pages

    def __len__(self):
        return self.pages

    def __str__(self):
        return self.title + " is written by " + self.author + " and has " + str(self.pages) + " pages."

    def __del__(self):
        print("Book Deleted")
```

```python
x=Books("Coding Ninjas", "Amit", 500)
```

```python
#Encapsulation..... we can hide an attribute

class Books():

    def __init__(self, title, author, pages):
        self.__title=title
        self.author=author
        self.pages=pages

    def __len__(self):
        return self.pages

    def __str__(self):
        return self.__title + " is written by " + self.author + " and has " + str(self.pages) + " pages."

    def __del__(self):
        print("Book Deleted")
```
